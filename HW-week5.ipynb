```python
import pandas as pd
from openpyxl import load_workbook

# è¯»å– Excel æ–‡ä»¶
df = pd.read_excel('D:/python/jupyter/ç ”äºŒ/module2-python/2023-Python-Basic/Week 5/H1.xls')
df.to_excel('D:/python/jupyter/ç ”äºŒ/module2-python/2023-Python-Basic/Week 5/H1.xlsx', index=False)
# # åŠ è½½ Excel æ–‡ä»¶
wb = load_workbook(filename='D:/python/jupyter/ç ”äºŒ/module2-python/2023-Python-Basic/Week 5/H1.xlsx')

# # è·å–ç¬¬ä¸€ä¸ªå·¥ä½œè¡¨
ws = wb.active

# åˆ›å»ºä¸€ä¸ªç©ºåˆ—è¡¨æ¥å­˜å‚¨è¶…é“¾æ¥
hyperlinks = []

# éå†å·¥ä½œè¡¨ä¸­çš„æ‰€æœ‰å•å…ƒæ ¼
for row in ws.iter_rows():
    for cell in row:
        # å¦‚æœå•å…ƒæ ¼æœ‰è¶…é“¾æ¥ï¼Œå°±å°†å®ƒæ·»åŠ åˆ°åˆ—è¡¨ä¸­
        if cell.hyperlink:
            hyperlinks.append(cell.hyperlink.target)

# å°†è¶…é“¾æ¥åˆ—è¡¨æ·»åŠ åˆ° DataFrame ä¸­
df['Hyperlinks2'] = pd.Series(hyperlinks)
df['Hyperlinks1'] = df['åœ¨çº¿æ–‡æ¡£åœ°å€ï¼ˆå¿…å¡«ï¼‰'].str.extract('(https?://[^\s]+)')
df['Hyperlinks1'] = df['Hyperlinks1'].fillna(df['Hyperlinks2'])
df = df.drop('Hyperlinks2', axis=1)
# å°†Hyperlinks1åˆ—é‡å‘½åä¸ºHyperlinks
df = df.rename(columns={'Hyperlinks1': 'Hyperlinks'})
df
```
