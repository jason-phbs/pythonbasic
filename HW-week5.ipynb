```python
import pandas as pd
from openpyxl import load_workbook

# ËØªÂèñ Excel Êñá‰ª∂
df = pd.read_excel('D:/python/jupyter/Á†î‰∫å/module2-python/2023-Python-Basic/Week 5/H1.xls')
df.to_excel('D:/python/jupyter/Á†î‰∫å/module2-python/2023-Python-Basic/Week 5/H1.xlsx', index=False)
# # Âä†ËΩΩ Excel Êñá‰ª∂
wb = load_workbook(filename='D:/python/jupyter/Á†î‰∫å/module2-python/2023-Python-Basic/Week 5/H1.xlsx')

# # Ëé∑ÂèñÁ¨¨‰∏Ä‰∏™Â∑•‰ΩúË°®
ws = wb.active

# ÂàõÂª∫‰∏Ä‰∏™Á©∫ÂàóË°®Êù•Â≠òÂÇ®Ë∂ÖÈìæÊé•
hyperlinks = []

# ÈÅçÂéÜÂ∑•‰ΩúË°®‰∏≠ÁöÑÊâÄÊúâÂçïÂÖÉÊ†º
for row in ws.iter_rows():
    for cell in row:
        # Â¶ÇÊûúÂçïÂÖÉÊ†ºÊúâË∂ÖÈìæÊé•ÔºåÂ∞±Â∞ÜÂÆÉÊ∑ªÂä†Âà∞ÂàóË°®‰∏≠
        if cell.hyperlink:
            hyperlinks.append(cell.hyperlink.target)

# Â∞ÜË∂ÖÈìæÊé•ÂàóË°®Ê∑ªÂä†Âà∞ DataFrame ‰∏≠
df['Hyperlinks2'] = pd.Series(hyperlinks)
df['Hyperlinks1'] = df['Âú®Á∫øÊñáÊ°£Âú∞ÂùÄÔºàÂøÖÂ°´Ôºâ'].str.extract('(https?://[^\s]+)')
df['Hyperlinks1'] = df['Hyperlinks1'].fillna(df['Hyperlinks2'])
df = df.drop('Hyperlinks2', axis=1)
# Â∞ÜHyperlinks1ÂàóÈáçÂëΩÂêç‰∏∫Hyperlinks
df = df.rename(columns={'Hyperlinks1': 'Hyperlinks'})
df
```
