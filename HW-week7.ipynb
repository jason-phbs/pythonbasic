#### Given a set of blue points, calculate the parameters of the fitted quadratic curve using the gradient descent method.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-1, 1, 100)
y = 1.5*x*x + 0.2 + + 0.2*np.random.randn(100)
plt.scatter(x, y)
plt.plot(x, 1.5*x*x + 0.2, 'r-', lw=3)
plt.show()
```

```python
import numpy as np
# 初始化参数
a = b = c = 0
# 学习率
lr = 0.1
# 迭代次数
epochs = 1000
# 梯度下降法
for i in range(epochs):
    y_pred = a*x*x + b*x + c
    grad_a = (-2/len(x)) * sum(x*x * (y - y_pred))
    grad_b = (-2/len(x)) * sum(x * (y - y_pred))
    grad_c = (-2/len(x)) * sum(y - y_pred)
    a = a - lr * grad_a
    b = b - lr * grad_b
    c = c - lr * grad_c
print(f"The parameters are: a={a}, b={b}, c={c}")
```
